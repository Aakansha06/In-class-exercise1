```python
# Please write your code here
import bs4
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
import bs4
import pandas as pd
import requests

a = []
def get_page_contents(url):
        page = requests.get(url, headers={"Accept-Language": "en-US"})
        return bs4.BeautifulSoup(page.text, "html.parser")
for i in range(0,500,10):
    url = 'https://www.yelp.com/search?find_desc=&find_loc=Mi&start='+str(i)
    print(url)
    soup = get_page_contents(url)
    Restaurants = soup.findAll('h4',class_='lemon--h4__09f24__1yd__ heading--h4__09f24__2ijYq alternate__09f24__39r7c')
    Res_Name = [Rest.find('a').text for Rest in Restaurants]
    a.extend(Res_Name)
for restaurant in Restaurants:
    removing_white_spaces = restaurant.a.strip()
    a.append(removing_white_spaces.replace('\n', ''))
```

```python
Restaurants_df = pd.DataFrame((a), columns =['Restaurants in Yelp'])
Restaurants_df
```

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from textblob import TextBlob
from nltk.stem import PorterStemmer
from textblob import Word
nltk.download('wordnet')
from nltk.tokenize import word_tokenize
from collections import Counter
```

```python
# Tokenization
Restaurants_df['Tokenization'] = Restaurants_df['Restaurants in Yelp'].apply(lambda x: TextBlob(x).words)
# Stemming
st = PorterStemmer()
Restaurants_df['Stemming'] = Restaurants_df['Tokenization'].apply(lambda x: " ".join([st.stem(word) for word in x]))
# Lemmatization
Restaurants_df['Lemmatization'] = Restaurants_df['Stemming'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
Restaurants_df
```

```python
names_of_restaurant = []
names = []
texts = Restaurants_df['Lemmatization'].tolist()
for name in texts:
  names_of_restaurant.append(nltk.pos_tag(word_tokenize(name)))
names_of_restaurant
```

```python
for name in names_of_restaurant:
  names.append(Counter(tag for word, tag in name))
names
```

```python
adj = []
adv = []
noun = []
poss = []
prep = []
prn = []
interj = []
verb = []
coordinating_conjunction = []
determiner = []
adjective = adverb = nou = possessor = preposition = pronoun = interjection = ver = cordcon = detr = 0
for elements in names_of_restaurant:
  for value in elements:
    if(value[1]=='JJ'):
      adjective = adjective+1
    elif (value[1]=='RB'):
      adverb = adverb+1
    elif (value[1]=='NN'):
      nou = nou+1
    elif (value[1]=='JJS'):
      possessor = possessor+1
    elif (value[1]=='IN'):
      preposition = preposition+1
    elif (value[1]=='NNS'):
      pronoun = pronoun+1
    elif (value[1]=='MD'):
      interjection = interjection+1
    elif (value[1]=='VB'):
      ver = ver+1
    elif (value[1]=='CC'):
      cordcon = cordcon+1
    elif (value[1]=='DT'):
      detr = detr+1
  adj.append(adjective)
  adv.append(adverb)
  noun.append(nou)
  poss.append(posessor)
  prep.append(preposition)
  prn.append(pronoun)
  interj.append(interjection)
  verb.append(ver)
  coordinating_conjunction.append(cordcon)
  determiner.append(detr)
  adjective = adverb = nou = possessor = preposition = pronoun = interjection = ver = cordcon = detr = 0
```

```python
df_pos_features = pd.DataFrame([adj, adv, noun, poss, prep, prn, interj, verb, coordinating_conjunction, determiner])
df_pos_features = df_pos_features.transpose()
df_pos_features.columns = ['Adjective', 'Adverb', 'Noun', 'Possessor', 'Preposition', 'Pronoun', 'Interjection', 'Verb', 'Coordinating Conjunction', 'Determiner']
df_pos_features.head(15)
```

```python
POS_df = pd.concat([Restaurants_df['Restaurants in Yelp'], df_pos_features],ignore_index=True, sort=False, axis=1)
POS_df.columns = ['Sentences', 'Adj', 'Adv', 'Noun', 'Poss', 'Prep', 'Prn', 'Interj', 'Ver', 'Coordinating Conjunction', 'Determiner']
POS_df
```
